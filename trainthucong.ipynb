```python
wines = pd.read_csv('D:\Machinelearning/train.csv')
wines.drop('residual sugar', axis=1, inplace=True)
wines.drop('free sulfur dioxide', axis=1, inplace=True)
wines.drop('total sulfur dioxide', axis=1, inplace=True)
wines
```

```python
wines = wines.rename(columns={'fixed acidity': 'fixed', 'volatile acidity': 'volatile','citric acid': 'citric', 'residual sugar': 'residual','free sulfur dioxide': 'free', 'total sulfur dioxide': 'total'})
```

```python
wines
```

```python
wines['type'] = wines['type'].map({'white': 0,'red': 1})
```

```python
Y = wines['quality']
X = wines[['fixed','volatile','citric','residual','chlorides','free','total','density','pH'
         ,'sulphates','alcohol','type']]
```

```python
X1 = sm.add_constant(X)
X1.head()
```

```python
results_mul = sm.OLS(Y,X1).fit()
```
