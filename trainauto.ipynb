```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
h2o_frame = h2o.H2OFrame(wines)
x = h2o_frame.columns
y = 'quality'
x.remove(y)
```

```python
h2o_automl = H2OAutoML(sort_metric='mse', max_runtime_secs=12*60, seed=666)
h2o_automl.train(x=x, y=y, training_frame=h2o_frame)
```

```python
h2o_models = h2o.automl.get_leaderboard(h2o_automl, extra_columns = "ALL")
h2o_models
```
